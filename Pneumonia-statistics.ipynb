```python
data_2018_onwards.plot(figsize = (13, 8), subplots=True, legend=True);
```

```python
year_groups = data_2018_onwards.groupby(pd.Grouper(freq= 'A'))
```

```python
data_2018_onwards_annual = pd.DataFrame()

for yr, group in year_groups:
    data_2018_onwards_annual[yr.year] = group.values.ravel()
    
```
